```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Gonzalez_images/fingerprint.tif', 0)
```

```python
kernel = np.array([[1,1,1],
                   [1,1,1],
                   [1,1,1]])
```

```python
eroded_img = cv2.erode(img, kernel, iterations=1)
```

```python
opening_img = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
```

```python
dil_op = cv2.dilate(opening_img, kernel, iterations=1)
```

```python
close_op = cv2.morphologyEx(dil_op, cv2.MORPH_CLOSE, kernel)
```

```python
import matplotlib.pyplot as plt

images = [img, eroded_img, opening_img, dil_op, close_op]
captions = ["Noisy Image", "Eroded", "Opening", "Dilation of opening", "Closing of opening"]

fig = plt.figure(figsize=(12, 8))

for i in range(5):
    ax = fig.add_subplot(2, 3, i + 1)
    ax.imshow(images[i], cmap='gray')
    ax.set_title(captions[i])
    ax.axis('off')
plt.tight_layout()
plt.show()
```
