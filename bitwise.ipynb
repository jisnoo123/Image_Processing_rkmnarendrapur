# Bitwise Operations

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Read the two images
img1 = cv2.imread('misc/4.1.04.tiff', 0)
plt.imshow(img1, cmap='gray')
```

```python
img2 = cv2.imread('misc/4.1.03.tiff', 0)
plt.imshow(img2, cmap='gray')
```

### AND Operation

```python
res_img = cv2.bitwise_and(img1, img2)
plt.imshow(res_img, cmap='gray')
```

### OR operation

```python
res_img = cv2.bitwise_or(img1, img2)
plt.imshow(res_img, cmap='gray')
```

### XOR operation

```python
res_img = cv2.bitwise_xor(img1, img2)
plt.imshow(res_img, cmap='gray')
```
