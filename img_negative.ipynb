# Negative Image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("./misc/house.tiff")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img.shape
```

```python
img_plot = plt.imshow(img)
```

```python
print(img)
```

```python
# Negative image
neg_img = np.zeros(shape=img.shape)

for channel in range(3):
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            neg_img[i][j][channel] = 255 - img[i][j][channel]
```

```python
neg_img = np.array(neg_img, np.int32)
print(neg_img)
```

```python
neg_img_plot = plt.imshow(neg_img)
```
