# Cropping

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def crop(img, piece, width, height):
    x_centre, y_centre = img.shape[0]//2, img.shape[1]//2
    cropped_img = img.copy()
    placed_img = cv2.resize(piece, (2*width, 2*height))
    cropped_img[x_centre-width:x_centre+width, y_centre-width:y_centre+width] = placed_img
    return cropped_img
```

```python
# Read the image
img = cv2.imread('misc/4.1.04.tiff', 0)
# Display the iamge
plt.imshow(img, cmap='gray')
```

```python
# Read the image to be placed
placed_img = cv2.imread('misc/4.1.03.tiff', 0)
# Display the iamge
plt.imshow(placed_img, cmap='gray')
```

```python
cropped_img = crop(img, placed_img, 50, 50)
plt.imshow(cropped_img, cmap='gray')
```
