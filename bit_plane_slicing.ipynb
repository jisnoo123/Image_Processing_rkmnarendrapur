# Bit Plane Slicing ###

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import math
```

```python
img = cv2.imread('misc/4.1.04.tiff', 0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
bit_len = int(math.log(img.shape[0], 2))
bit_plane_imgs = list()

for plane in range(bit_len):
    bit_pl_img = np.zeros(img.shape)
    for i in range(img.shape[0]):
        for j in range(img.shape[0]):
            bit_pl_img[i][j] = int(format(img[i][j], '08b')[plane])
    bit_plane_imgs.append(bit_pl_img)
```

```python
plt.figure(figsize=(12, 6))

for i in range(len(bit_plane_imgs)):
    plt.subplot(2, 4, i + 1)
    plt.imshow(bit_plane_imgs[i], cmap='gray')
    plt.title(f'Bit Plane {7-i}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
