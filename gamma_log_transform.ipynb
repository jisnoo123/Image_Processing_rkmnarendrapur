```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
img = cv2.imread("./misc/house.tiff")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_plot = plt.imshow(img)
print(img)
```

```python
bw_img = cv2.imread("./misc/house.tiff", 0)

gamma_img = np.int64(np.power(bw_img, 0.25))
```

```python
gamma_plot = plt.imshow(gamma_img, cmap='gray')

print(gamma_img)
```

```python
# Log transform
log_tr_img = np.zeros(shape=bw_img.shape)


for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        pixel = gamma_img[i][j]
        #print(pixel)
        #print(math.log(1+img[i][j][channel]))
        #if pixel == 255:
            #pixel = 255
        log_tr_img[i][j] = int( math.log(0.001 + pixel) )
          
```

```python
log_tr_img = np.array(log_tr_img, np.uint8)
print(log_tr_img)
```

```python
log_tr_img = np.array(log_tr_img, np.uint32)
log_tr_img_plot = plt.imshow(log_tr_img, cmap='gray')
```
