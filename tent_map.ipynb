# Tent Map Randomization

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

### Reading the image

```python
# Read the image
img = cv2.imread('misc/4.1.04.tiff', 0)
# Display the iamge
plt.imshow(img, cmap='gray')
```

### Random Numbers Generation

```python
def generate_rn(img):
    # Generates random numbers
    x_0 = 0.3
    h = 1.4
    x = list()
    x.append(x_0)
    for i in range(1, img.shape[0]**2):
        if x[i-1]<0.5:
            x_new = x[i-1]*h
        else:
            x_new = (1-x[i-1])*h
        x.append(x_new)
    return x
```

### Computation

```python
rn = generate_rn(img)
sorted_rn = rn.copy()
sorted_rn.sort()
indices = [None]*(img.shape[0]**2)
```

```python
for i in range(len(rn)):
    for j in range(len(sorted_rn)):
        if rn[i] == sorted_rn[j]:
            indices[i] = j
            break
```

```python
img_cpy = img.copy()
img_cpy = img_cpy.flatten()

new_img = np.zeros(img.shape[0]**2)

index = 0
for i in indices:
    new_img[i] = img_cpy[index]
    index += 1
```

### Result

```python
# Display Tent Map image
new_img = new_img.reshape(256,256)
```

```python
plt.imshow(new_img, cmap='gray')
```
