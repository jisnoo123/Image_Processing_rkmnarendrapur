### Salt & Pepper Noise

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import random
```

```python
def salt_pepper(img, d):
    # d is the density
    locations = [coord for coord in np.ndindex(img.shape)]
    num_salt = num_pep = int(img.size*d) # No. of locations to be noised

    random_locs_salt = random.sample(locations, num_salt)
    random_locs_pep = random.sample(locations, num_pep)

    noised_img = img.copy()

    # Add noise to the img

    for locs in random_locs_salt:
        noised_img[locs[0]][locs[1]] = 255

    for locs in random_locs_pep:
        noised_img[locs[0]][locs[1]] = 0

    return noised_img
```

```python
# Read the image
img = cv2.imread('misc/4.1.04.tiff', 0)
# Display the iamge
plt.imshow(img, cmap='gray')
```

```python
# Apply Salt and Pepper Noise
new_img = salt_pepper(img, 0.2)

plt.imshow(new_img, cmap='gray')
```
