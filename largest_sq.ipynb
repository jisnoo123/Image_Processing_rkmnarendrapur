# Largest Square Problem

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

### Read the image

```python
img = cv2.imread('Gonzalez_images/square.tif', 0)
```

### Computation

```python
kernel = np.ones((13,13))
```

```python
eroded_img = cv2.erode(img, kernel, iterations=1)
```

```python
dilated_img = cv2.dilate(eroded_img, kernel, iterations=1)
```

### Display results

```python
images = [img, eroded_img, dilated_img]
captions = ["All squares", "Eroded image", "Dilated with largest squares only"]

fig = plt.figure(figsize=(12, 8))

for i in range(3):
    ax = fig.add_subplot(1, 3, i + 1)
    ax.imshow(images[i], cmap='gray')
    ax.set_title(captions[i])
    ax.axis('off')
plt.tight_layout()
plt.show()
```
