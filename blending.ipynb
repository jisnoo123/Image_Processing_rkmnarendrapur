# Image Blending

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Read the two images
img1 = cv2.imread('misc/4.1.04.tiff')
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1_rgb)
```

```python
img2 = cv2.imread('misc/4.1.03.tiff')
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img2_rgb)
```

### Simple Blending

```python
res_simple_blending = cv2.add(img1, img2)
res_simple_blending_rgb = cv2.cvtColor(res_simple_blending, cv2.COLOR_BGR2RGB)
plt.imshow(res_simple_blending_rgb)
```

### Weighted Blending

```python
res_weighted_blending = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)
res_weighted_blending_rgb = cv2.cvtColor(res_weighted_blending, cv2.COLOR_BGR2RGB)
plt.imshow(res_weighted_blending_rgb)
```

### Area wise image blending

```python
img2_gray = cv2.imread('misc/4.1.03.tiff', 0)
ret, mask = cv2.threshold(img2_gray, 128, 255, cv2.THRESH_BINARY)
portion1 = cv2.bitwise_and(img1, img1, mask = mask)
mask_inv = cv2.bitwise_not(mask)
portion2 = cv2.bitwise_and(img2, img2, mask = mask_inv)
final = cv2.bitwise_or(portion1, portion2)
```
