# Median Filtering

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

### Reading the image

```python
# Read the image
img = cv2.imread('misc/4.1.04.tiff', 0)
# Display the iamge
plt.imshow(img, cmap='gray')
```

### Padding the image

```python
# Pad the image
padded_img = np.pad(img, pad_width=1, mode='constant', constant_values=0)
print('Padded image')
plt.imshow(padded_img, cmap='gray')
```

### Generating Median Filtered Image

```python
new_img = np.zeros(img.shape)
```

```python
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        neighbours = [padded_img[i][j], padded_img[i][j+1], padded_img[i][j+2], padded_img[i+1,j], padded_img[i+1][j+1], padded_img[i+1][j+2], padded_img[i+2][j], padded_img[i+2][j+1], padded_img[i+2][j+2]]
        neighbours_arr = np.array(neighbours)
        new_img[i][j] = np.median(neighbours_arr)
```

```python
print('Median filtered image')
plt.imshow(new_img, cmap='gray')
```
